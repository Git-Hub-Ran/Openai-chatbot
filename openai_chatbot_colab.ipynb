My chatbot is a virtual Spanish tutor designed for beginners. Its purpose is to help users learn basic Spanish for everyday situations such as greetings, shopping, and simple conversations. The bot uses prompt engineering techniques such as role definition, contextual instructions, and constrained responses to ensure explanations are simple, clear, and suitable for language learners.

```python
#using Open AI:
import openai
from openai import OpenAI
import os

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

#This function sends a list of messages to the OpenAI ChatCompletion API and returns the generated text response from the model.
def get_completion_from_messages(messages, model="gpt-4o-mini", temperature=0):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
    #print(str(response.choices[0].message)) #print the entire message response to see the role and content
    return response.choices[0].message.content

#list if messages:
messages = [
    {"role":"system", "content": "you are a virtual Spanish tutor designed for beginners that is also funny and likes to tell jokes.ive the correct Spanish sentence the user should say."},#set the behavior of assistant
    {'role':'user', 'content':'I want to buy milk in the supermarket in Spain, how do I say it in Spanish?'},
    {"role":"assistant","content":"what kind of milk do you want to buy?"},
    {'role':'user', 'content':'Cow milk lactose free'}
]

response = get_completion_from_messages(messages,temperature=1)
print(response)


```
